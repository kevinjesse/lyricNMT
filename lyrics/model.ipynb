```python
import pandas as pd
import torch
import torch.nn as nn
import torch.utils.data as utils
import torch.nn.functional as F
import numpy as np
use_cuda = torch.cuda.is_available()
device = torch.device('cuda' if use_cuda else "cpu")
epochs = 100
df = pd.read_pickle('data.txt')
x = df['x'][0]
y = df['y'][0]
print(x.shape)
# print(x.shape)

class LSTM(torch.nn.Module):
    def __init__(self):
        super(LSTM, self).__init__()
        self.l1 = nn.LSTM(52,128, batch_first=True)
        self.lin = nn.Linear(128, 52)
            #nn.Softmax())
    def forward(self, x, hidden):
        out, hidden = self.l1(x, hidden)
        out = out.contiguous()
#         print(out.size(2))
#         print(out.size())
#         print(out.view(-1, out.size(2)).shape())
        out = self.lin(out.view(-1, out.size(2)))
        return F.softmax(out).view(-1,20,52)
```

```python
model = LSTM().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
train = utils.TensorDataset(torch.from_numpy(np.int64(x)),torch.from_numpy(np.int64(y)))
train_data = utils.DataLoader(train, batch_size=8)
```

```python
def train_NN(model):
    model.train()
    for inputs, labels in train_data:
        inputs = inputs.type(torch.FloatTensor)
        inputs = inputs.to(device)
        labels = labels.to(device)

    #     
        optimizer.zero_grad()
        h0 = torch.randn(1, 8, 128)
        c0 = torch.randn(1, 8, 128)
        outputs = model(inputs, (h0,c0))
        loss = criterion(outputs, labels)
        print(loss)
        loss.backward()
        optimizer.step()
    return model
    
```

```python

import time

start = time.time()
train_NN(model)
end = time.time()
print(end - start)
```
